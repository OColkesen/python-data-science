```python
#loading three important modules
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#load the data and use the .info() to look at the meta data
path = "http://csc110.drchesmith.com/City_Employee_Salaries_January_2019.csv"
clt = pd.read_csv(path)
clt.info()
```

```python
#display the first five rows of data
clt.head()
```

```python
#create a variable with all the counts for different FID values and create a new list to include all FIDs
#that exist more than once to see if FID actually refers to ID Number.
id_check = pd.value_counts(clt.FID)
id_check[id_check > 1]
```

```python
#quicker way to see if the number of IDs is the same as number of rows
clt.FID.nunique()
```

```python
temp_check = pd.value_counts(clt.Reg_Temp)
temp_percentage = 100*temp_check[1]/(temp_check.sum())
temp_percentage
```

```python
unit_check = pd.value_counts(clt.Unit)
unit_check[0:10]
#or unit_check.head(10)
```

```python
clt_mfs = clt[clt.Unit == "M&FS"]
pd.value_counts(clt_mfs.Dept)
```

```python
dept_check = pd.value_counts(clt.Dept)
dept_check[0:10]
```

```python
#first we group the data by department and unit while taking the counts
#.sort_values part lists the rows in descending order with respect to counts
#[[]] is used to just pick the Counts column to display from the nested list of data
dept_unit_list = clt.groupby(['Dept', 'Unit']).count()
dept_unit_list[["Counts"]].sort_values(by = "Counts", ascending = False)
```

```python
people_check = clt.sort_values(by = "Annual_Rt", ascending = False)
people_check
```
